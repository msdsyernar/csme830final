#### Missingness and Cardinality

```python
import missingno as msno
# 1. Select only the columns that have a numeric data type
df_numeric = df.select_dtypes(include=['number'])
df_numeric.isna().sum()
```

_**Cardinality** refers to the number of unique values in a column. For categorical features, high cardinality (many unique values) can complicate one-hot encoding, so it's essential to identify them._

```python
categorical_cols = df.select_dtypes(include='object').columns
cardinality_series = df[categorical_cols].nunique().sort_values(ascending = False)
cardinality_series
```

#### Correlation

```python
corr_matrix = df_numeric.corr()

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=False, cmap= 'coolwarm', fmt = '.2f', linewidths=.5, linecolor= 'black')
plt.title('Correlation Matrix of All Numeric Features')
plt.show()

target_corr = corr_matrix['SalePrice'].sort_values(ascending= False)
print(target_corr)
```

#### Normality

```python
# Creating a subplots to check the normality with Histogram and QQ plot
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (12, 5))

# Plot the histogram
sns.histplot(df['SalePrice'], kde= True, ax = ax1)
ax1.set_title('Sales Price Distribution')

# QQ plot
stats.probplot(df['SalePrice'], dist='norm', plot = ax2)
ax2.set_title('QQ plot of SalePrice')

plt.tight_layout()
plt.show()
```

```python
# Apply the log transformation to the SalePrice column
df['SalePrice_log'] = np.log1p(df['SalePrice'])
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (12, 5))

# Plot the histogram
sns.histplot(df['SalePrice_log'], kde= True, ax = ax1)
ax1.set_title('Sales Price Distribution')

# QQ plot
stats.probplot(df['SalePrice_log'], dist='norm', plot = ax2)
ax2.set_title('QQ plot of SalePrice')

plt.tight_layout()
plt.show()
```

Let's check for outliers. As previously, I have a target correlation. I will set a threshold and select important features to work on. 

```python
high_impact_features = target_corr[target_corr > 0.4].index.tolist()
print(f"These are the features which have high impact on the target variable:\n{high_impact_features}")
```

```python
outlier_features = high_impact_features

plt.figure(figsize=(25,15))

for i, feature in enumerate(outlier_features):
    plt.subplot(2, len(outlier_features)//2+1, i+1)
    sns.boxplot(y = df[feature])
    plt.title(f'{feature}')
plt.tight_layout()
plt.show()
```

**Summary of outliers:** 
- Outliers in SalePrice is understandable, as showed in the Distribution plot, the line was right skewed, which makes sense. 
- Overall Qual has a single outlier.
- Gr Liv Area has many outliers around 4000 - 5000 sqft. Above Ground Living Area
- Garage cars and Garage Area: Both show numerous outliers on the high side, indicating homes with significantly larger garage capacity or area than the typical house.
- Total Bsmt SF (Total Basement Square Footage): Shows several high outliers, meaning some homes have exceptionally large basements.
